#### Finn egenvektorene og egenverdiene til den nye kovariansmatrisen $\sigma_y $. Hva forteller disse oss om nivåkurvene til sannsynlighetstettheten av y.

$\Sigma_y = A \Sigma A^T$

$
\Sigma_y = \begin{bmatrix}
0.7 & 0.5 \\
0.5 & 0.5
\end{bmatrix}
$

```python
import numpy as np
from matplotlib.patches import Ellipse
import matplotlib.pyplot as plt
```

```python
cov_matrix_y = np.array([
    [0.7, 0.5],
    [0.5, 0.5]
])

egenverdier, egenvektorer= np.linalg.eig(cov_matrix_y)

print('Egenverdier:', egenverdier)
print('Egenvektorer:', egenvektorer)
```

```python
fig, ax = plt.subplots(figsize=(5,5))

for scale in [1,2,3]:
    a = np.sqrt(egenverdier[0]) * scale
    b = np.sqrt(egenverdier[1]) * scale

    angle = np.degrees(np.arctan2(egenvektorer[1,0], egenvektorer[0, 0]))

    elipse = Ellipse((0, 0), 2*a, 2*b, angle=angle, fill=False, edgecolor='blue', linewidth=2)

    ax.add_patch(elipse)

for i, (eigenvalue, eigenvector) in enumerate(zip(egenverdier, egenvektorer.T)):
   ax.arrow(0, 0, eigenvector[0]*np.sqrt(eigenvalue)*3, 
            eigenvector[1]*np.sqrt(eigenvalue)*3, 
            head_width=0.15, head_length=0.1, fc='red', ec='red')

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_xlabel('y1')
ax.set_ylabel('y2')
ax.set_title('Nivåkurver med egenvektorer')
plt.show()
```
