```python
import numpy as np
import matplotlib.pyplot as plt
import json
from scipy.sparse import load_npz
```

```python
# Laster inn matrise
A_sparse=load_npz('NMBU_adjacency_matrix.npz')

# Lager en dict for som gir url-en for hver matrise.
with open("adj_index.json", 'r') as f:
    index_to_url = json.load(f)


# Omgjør en sparse matrise til "vanlig" matrise A
A= A_sparse.toarray()
```

```python
# finn kolonnesummene i A
m = A.sum(0)
m
```

```python
# finn overgangsmatrisen P
P = A/m[None:]
P
```

```python
# beregn egenverdiene og egenvektroene til P. Sjekk at egenverdi1 = 1
egenverdier, egenvektorer = np.linalg.eig(P)
print(egenverdier[0])
```

```python
# finn hva den konverger mot

idx = np.argmax(np.real(egenverdier))

pi = np.real(egenvektorer[:,idx])

# normaliser
pi = pi / pi.sum()

print(pi)
```

```python
# finn websiden med høyest pagerank, høyest pi verdi
top_list = np.argsort(pi)[::-1]
for i in top_list:
    print(f'score: {pi[i]} | URL: {index_to_url[str (i)]}')
```
